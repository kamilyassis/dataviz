```python
ambulatorio = (
    ambulatorio
    .sort_values(by="nu_ocorrencias", ascending=False)
    .reset_index(drop=True)
)

ambulatorio
```

```python
chaves_internacoes = set(internacoes["no_municipio"].dropna().astype(str).unique())
chaves_ambulatorio = set(ambulatorio["no_municipio"].dropna().astype(str).unique())

chaves_comuns = sorted(chaves_internacoes & chaves_ambulatorio)

print(f"Total de chaves comuns: {len(chaves_comuns)}")
chaves_comuns
```

```python
dataset = (
    ambulatorio.loc[ambulatorio["no_municipio"].isin(chaves_comuns), ["co_ibge", "no_municipio", "nu_ocorrencias", "mp_municipio"]]
    .rename(columns={"nu_ocorrencias": "nu_ocorrencias_ambulatorio"})
    .merge(
        internacoes.loc[internacoes["no_municipio"].isin(chaves_comuns), ["co_ibge", "no_municipio", "nu_ocorrencias"]]
        .rename(columns={"nu_ocorrencias": "nu_ocorrencias_internacoes"}),
        on=["co_ibge", "no_municipio"],
        how="left"
    )
)

dataset["nu_ocorrencias_internacoes"] = dataset["nu_ocorrencias_internacoes"].fillna(0).astype(int)
dataset = dataset.sort_values("no_municipio").reset_index(drop=True)

dataset
```

```python
mapa_uf = {
    "11":"RO","12":"AC","13":"AM","14":"RR","15":"PA","16":"AP","17":"TO",
    "21":"MA","22":"PI","23":"CE","24":"RN","25":"PB","26":"PE","27":"AL","28":"SE","29":"BA",
    "31":"MG","32":"ES","33":"RJ","35":"SP",
    "41":"PR","42":"SC","43":"RS",
    "50":"MS","51":"MT","52":"GO","53":"DF"
}

dataset["uf_cod"] = dataset["co_ibge"].astype(str).str.zfill(6).str[:2]
dataset["uf"] = dataset["uf_cod"].map(mapa_uf)
dataset = dataset.drop(columns=["uf_cod"])
```

```python
dataset = dataset[dataset["nu_ocorrencias_ambulatorio"] != 0]
dataset = dataset[dataset["nu_ocorrencias_internacoes"] != 0]
dataset
```

```python
dataset = dataset.sort_values(by="nu_ocorrencias_internacoes", ascending=False).reset_index(drop=True)
dataset
```
